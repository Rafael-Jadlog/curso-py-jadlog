```python
# Entrada de dados
nome = str(input("Nome: ")).title()
cargo = int(input("1.Analista Financeiro | 2.Assistente Financeiro | 3.Coordenador >>> "))
salario = float(input("Salario: "))
tempo_casa = int(input("Tempo de Casa: "))
 
```

```python
match cargo:
    case 1:  # Analista
        cargo = 'Analista Financeiro'
    case 2:  # Assistente Financeiro
        cargo = 'Assistente Financeiro'
    case 3:  # Coordenador
        cargo = 'Coordenador'
    case _:
        cargo = "Op√ß√£o inv√°lida!"
```

```python
# Classifica√ß√£o do funcion√°rio por tempo de experi√™ncia

if tempo_casa <= 2 and tempo_casa >=0:  #and & or |
    senioridade = 'J√∫nior'
elif tempo_casa <= 5 and tempo_casa >=3:
    senioridade = 'Pleno'
elif tempo_casa >= 6 and tempo_casa <=40:
    senioridade = 'S√™nior'    
else:
    senioridade= 'erro'
```

```python
# Aplica√ß√£o de l√≥gica de decis√£o 

if cargo == 'Analista Financeiro' and senioridade != 'J√∫nior':                       #!= diferente
    permissao = 'Gerenciamento Liberado'

elif cargo == 'Assistente Financeiro' and senioridade == 'J√∫nior':                       #!= diferente
    permissao = 'Apenas operacoes basicas'    

elif cargo == 'Coordenador' and senioridade == 'S√™nior':                       #!= diferente
    permissao = 'Acesso total'

elif senioridade == 'erro':
    permissao = 'Verifique as respostas anteriores'
    
else:
    permissao = 'Permissao negada'
permissao
```
