```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)

print(a / b)    # 10 / 3 = 3.333333
print(a // b)   # 10 / 3 = 3
print(a % b)    # 10 / 3 = 3 * 3 = 9

print(a ** b)   # 10³'
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace">
    <strong></br></br></br></br>
        Estrutura condicional IF
    </strong>
</p>

```python
# Condição simples - Status por avalição

avaliação = 4.2

if avaliação >= 7.0:  # True
    print("Legal, obrigado por sua avaliação!")

else:
    print("Que pena! Vamos melhorar...")   
```

```python
# Condição composta

avaliação = 10

if avaliação >= 7.0:  # True
    print("Legal, obrigado por sua avaliação!")

elif avaliação >= 5.0:
    print("Legal, faremos alguns ajustes!")

else:
    print("Que pena! Vamos melhorar...")   
```

```python
# Condição aninhada

avaliação = 6.0

if avaliação >= 7.0:  # True
    print("Legal, obrigado por sua avaliação!")

elif avaliação >= 5.0:
    resposta = input("Quer justificar sua resposta? [S ou N]: ").upper()

    if resposta == "S":
        print("ok, então justifique: ")
    elif resposta == "N":
        print("Hunmn, ta com pressa é...")
    else:
        print("Resposta inválida!")

else:
    print("Que pena! Vamos melhorar...")   
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace">
    <strong></br></br></br></br>
        Match/Case
    </strong>
</p>

```python
menu = int(input("1•Cadastrar | 2•Visualizar | Opção: "))

match menu:
    case 1:
        print("escolheu 1")
    case 2:
        print("escolheu 2")
    case _:
        print('Resposta inválida!')
```
