```python
import pandas as pd
```

CRIANDO DATAFRAME

```python

df_entregas = pd.DataFrame({
    'id_entrega': [1, 2, 3],                     # chave primaria
    'filial':     ['RJ', 'MG', 'SP'],
    'status':     ['ENTREGUE', 'ATRASO', 'ENTREGUE']  
})

df_custos = pd.DataFrame({
    'id_custo':   [1, 2, 3],   
    'id_entrega': [1, 1, 3],
    'custo':      [150.09, 147.56, 278.26]
})
```

JOIN

```python
#inner join

pd.merge(
    df_entregas,df_custos,
    on='id_entrega',
    how='outer'
)
```

```python
df_entregas_idx = df_entregas.set_index('id_entrega')
df_custos_idx = df_custos.set_index('id_entrega')   

#para relacionar com join e necessario criar index pois e atraves de numeracao que se localizam

df_join = df_entregas_idx.join(df_custos_idx,how='left')
df_join.reset_index()
```
