```python
senha_cadastrada = input('Senha: ')
senha_entrada = ''
tentativas = 0

while  senha_cadastrada != senha_entrada :
    senha_entrada = input('Digite sua senha: ')
    tentativas += 1

if tentativas == 3 and senha_cadastrada != senha_entrada:
   print ('Tentativas maximas atingidas')
   break
else:
    print('Liberado')       
```

```python
# Entrada de valores

franquias = ['RJ Rio de Janeiro 55','LJ Sao Paulo 55','LJ Tres Rios 11']

pg_franquias = []

for franquia in franquias:
    valor = float(input(f'{franquia} | Valor: '))
    
    pg_franquias.append(franquia)
    pg_franquias.append(valor)

pg_franquias    


```

```python
# Cálculo

soma = 0
for indice, dado in enumerate(pg_franquias):
    if indice%2 ==1:
        soma += dado *0.1


print(soma)

```

```python
franquiass = ['RJ Rio de Janeiro 55', 'LJ Sao Paulo 55', 'LJ Tres Rios 11']

aliquota = 0.10
total_royalties = 0.0

for item in franquiass:
    partes = item.split()
    codigo = partes[0]
    cidade = " ".join(partes[1:-1])  # ignoramos o número que veio na lista

    # pede o faturamento ao usuário
    entrada = input(f"Digite o faturamento de '{codigo} - {cidade}': ").strip()
    # permite vírgula decimal
    entrada = entrada.replace(",", ".")
    faturamento = float(entrada)

    royalty = faturamento * aliquota
    total_royalties += royalty

    print(f"{codigo} - {cidade} | Faturamento: {faturamento:.2f} | Royalty (10%): {royalty:.2f}")

print(f"\nTotal de royalties a receber: {total_royalties:.2f}")
```

```python
franquias_cadastradas =[]

for contador in range(2):

    franquia = dict(
        id            = int(input('ID: ')),
        cpf           = input('CPF: '),
        endereco      = input('ENDERECO: ').title(),
        bairro        = input('BAIRRO: ').title(),
        cidade        = input('CIDADE: ').title(),
        estado        = input('ESTADO: ').upper()
    )

    franquias_cadastradas.append(franquia)

franquias_cadastradas    
```

```python


conta_unidades = 0   

for item in franquias_cadastradas:
    conta_unidades += 1   

    print(f"{conta_unidades}° franquia")
    print(f"ID = {item['id']}")
    print(f"CPF = {item['cpf']}")
    print(f"Bairro = {item['bairro']}")
    print(f"Endereço = {item['endereco']}")
    print(f"Cidade = {item['cidade']}")
    print(f"Estado = {item['estado']}")
    print()
    

```
