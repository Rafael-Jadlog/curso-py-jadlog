```python
import pandas as pd

produtos= [
    ['lapis','Escola', 1, 10],
    ['caneta','Escola', 1.5, 20],
    ['borracha','Escola', 0.5, 50],
    ['mochila','Escola', 22, 5],
    ['caderno','Escola', 5, 2],
    ['estojo','Escola', 3, 15],
    ['arroz','mercado', 20, 10],
    ['feijao','mercado', 10, 5],
    ['feijao','mercado', 10, 2]
    ]

colunas = ['Produto','Categoria', 'Valor', 'Qtde']

df = pd.DataFrame(produtos, columns=colunas)

df.head() # 5 primeiras linhas e 5 últimas linhas
df.shape # qtd linhas e colunas
df.info() # resumo do dataframe
df.describe() # estatísticas descritivas
df.dropna() # remover linhas com valores nulos
# strip() # remover espaços em branco precisa ser feito dentro de um loop ou usando apply
df['Produto']

df.groupby('Produto')['Qtde'].mean()
```

```python
df[df['Valor']>0]

df.describe()

df['Valor'].max()

df.groupby('Categoria')['Valor'].max()

df.groupby('Produto').agg(
   Valor_unit= ('Valor','max'),
   total_em_estoque= ('Qtde','sum'),
)
```

```python

# Filtrar apenas valores > 0 (se fizer sentido no seu caso)
df_pos = df[df['Valor'] > 0].copy()

# Criar a coluna Valor_total linha a linha (Valor * Qtde)
df_pos['Valor_total'] = df_pos['Valor'] * df_pos['Qtde']

# Agregar por Produto
resumo = df_pos.groupby('Produto').agg(
    Valor_unit     = ('Valor', 'max'),   # maior preço unitário do produto
    total_em_estoque   = ('Qtde',  'sum'),   # soma das quantidades
    Valor_total = ('Valor_total', 'sum')  # soma de (Valor*Qtde) das linhas
).reset_index()

resumo
```
