```python
#nome_arquivo = 'base_entregas_brutas'

def ler_arquivo(nome_arquivo):
    try:
        df = pd.read_excel(f'dados/{nome_arquivo}.xlsx')
        return df
    except FileNotFoundError as e:
        print (f'Erro:{e}')

def criar_atualizar_arquivo(nome_arquivo,df_tratado = ''):
    df_tratado.to_excel(f'dados/{nome_arquivo}.xlsx'
)            


df =ler_arquivo('base_entregas_brutas')

# controle_path = 'dados/base_entregas_brutas.xlsx'
# try: 
#     df = pd.read_excel(controle_path)
# except FileNotFoundError as e:
#     print(e)
```

<p style="font-size: 1.1rem; color: #9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Dados ausentes        
    </strong>
</p>

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Por segurança, criamos uma copia do DF       
    </strong>
</p>

```python
df_tratado = df.copy()

df_tratado.head()
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Filial 
    </strong>
</p>

```python
#df_tratado.loc[2]

df_tratado.loc[:,'filial'] = df_tratado['filial'].fillna('Não informada')
df_tratado.loc[:,'descricao_entrega'] = df_tratado['descricao_entrega'].fillna('ok')
df_tratado.loc[:,'tipo_servico'] = df_tratado['tipo_servico'].fillna('validar')
df_tratado.loc[:,'status'] = df_tratado['status'].fillna('Não informada')

#df_tratado.loc[:,'prazo_previsto'] = df_tratado['prazo_previsto'].fillna('validar')





df_tratado.sample(5)

```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Cliente 
    </strong>
</p>

```python
df_tratado.loc[:,'cliente'] = df_tratado['cliente'].fillna('outros')
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Dias de atraso 
    </strong>
</p>

```python
df_tratado.loc[:,'dias_atraso'] = df_tratado['dias_atraso'].fillna(0) #fillna('=2+2')
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Prazo previsto 
    </strong>
</p>

```python
df_tratado.loc[:,'prazo_previsto'] = df_tratado['prazo_previsto'].fillna(
    df_tratado['prazo_previsto'].median()
)

df_tratado.head(7)
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Duplicadas
    </strong>
</p>

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Verificando duplicadas com base no ID 
    </strong>
</p>

```python
df_tratado.duplicated(subset=['id_entrega']).sum()
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Removendo duplicadas, mantendo a primeira ocorrencia
    </strong>
</p>

```python
#df_tratado.drop_duplicates(subset=['id_entrega'],inplace=True) esta jeito com inplace sera descontinuado!!!!

df_tratado = df_tratado.drop_duplicates(subset=['id_entrega'])

df_tratado.duplicated(subset=['id_entrega']).sum()
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Tipagem correta para ID
    </strong>
</p>

```python
df_tratado['id_entrega'] = df_tratado['id_entrega'].astype('Int64')
df_tratado['prazo_previsto'] = df_tratado['prazo_previsto'].astype('int64')
df_tratado['dias_atraso'] = df_tratado['dias_atraso'].astype('int64')

df_tratado.dtypes
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Tratamento - Padronização e Normalização
    </strong>
</p>

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Removendo espaços e alterando para caixa alta 
    </strong>
</p>

```python
df_tratado['status'] = df_tratado['status'].str.strip().str.upper()
df_tratado['tipo_servico'] = df_tratado['tipo_servico'].str.strip().str.upper()

df_tratado
```

<p style="font-size: 1.1rem; color: #00d9ff; font-family: monospace ">
    <strong>
        Status 
    </strong>
</p>

```python
mapa_status = {
    'ENTREGUES'    : 'ENTREGUE',
    'EM TRÂNSITO' : 'EM TRANSITO',
    'EM TRANZITO' : 'EM TRANSITO',
    'ATRAZADO'    : 'ATRASADO',
    'ATRASADO'    : 'ATRASADO',
    ' ATRASADO'    : 'ATRASADO',
    'ENTREGUE'    : 'ENTREGUE',
    'EM TRÂNSITO' : 'EM TRANSITO',
    'NÃO INFORMADA' : 'NAO INFORMADA', #Não informada
    'Entregue ': 'ENTREGUE',
    'entregue': 'ENTREGUE',
    'ATRASADO': 'ATRASADO',
    'EM TRANSITO': 'EM TRANSITO',
    'Em trânsito': 'EM TRANSITO',
}

df_tratado['status'] = df_tratado['status'].map(mapa_status)

df_tratado.head(5)

```

CRIANDO COLUNA CALCULADA

```python
df_tratado['data_prevista'] = (
    df_tratado['data_postagem'] + 
    pd.to_timedelta(df_tratado['prazo_previsto'],unit = 'D')  #(d = dia) (m = mes) (y = ano)
)

df_tratado[
    [
    'data_postagem',
    'prazo_previsto',
    'data_prevista'
    ]
]

df_tratado.head(2)
```

```python
hoje = pd.Timestamp.today().normalize()

df_tratado['Atraso']=(
    hoje - df_tratado['data_prevista']
).dt.days
df_tratado['Atraso'] = df_tratado['Atraso'].clip(lower=0)

df_tratado
```

```python
df_tratado['Atraso'] = (df_tratado['dias_atraso'] > 0).astype(int)

print('qtde_atraso')
print(df_tratado['dias_atraso'].sum())
df_tratado.head(5)
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Prazo Previsto de acordo com a filial

```python
df_tratado.loc[
    (df_tratado['cliente']=='Shopee') & (df_tratado['filial']=='RJ'),
    'prazo_previsto'
] = 5
df_tratado.loc[
(df_tratado['cliente']=='Amazon') & (df_tratado['filial'].isin(['MG','SP'])),
    'prazo_previsto'
] = 2


df_tratado
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/><br/><br/>
        Atualizando arquivo para executar formulas
    </strong>
</p>

```python
app = xw.App(visible = True)  #abre o arquivo excel e deixa tela oculta

wb = xw.Book('dados/base_entregas_tratadasx.xlsx') #abre o caminho do arquivo

wb.app.calculate()  #forca calcular as formulas da planilha

wb.save() #Atualiza o aqruivo

wb.close() #Fecha o arquivo

app.quit() #Fecha o aplicativo excel
```

```python
criar_atualizar_arquivo('base_entregas_tratadasx',df_tratado=df_tratado)
```
