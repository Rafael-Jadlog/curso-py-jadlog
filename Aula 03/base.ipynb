```python
# Quantas contas estão vencidas?

from datetime import datetime

hoje = datetime.today()
contas_vencidos = []
contas_vencidos_receber=[]

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == 'Em aberto' and conta['tipo'] == 'Pagar':
        contas_vencidos.append(conta)

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == 'Em aberto' and conta['tipo'] == 'Receber':
        contas_vencidos_receber.append(conta)        

print(f'qtde de boletos a pagar em atraso {len(contas_vencidos)}')
print(f'qtde de boletos a receber em atraso {len(contas_vencidos_receber)}')

```

```python
# Total em contas vencidas?

total_vencido = 0

for conta in contas_vencidos:
    #total_vencido = total_vencido + conta['valor']
    total_vencido += conta['valor']

round(total_vencido,2)    


```
