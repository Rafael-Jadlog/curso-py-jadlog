IMPORTAR BIBLIOTECAS

```python
from datetime import datetime
import json
```

CAMINHO DO ARQUIVO

```python
try:
    with open(r'dados\contas.json','r',encoding='UTF8') as arquivo:
        contas = json.load(arquivo)

except FileNotFoundError as e:
    print(e)
    
#finally:print('batata')

contas
```

TRATAMENTO DE DADOS

```python
contas
for conta in contas:
    conta['vencimento'] = datetime.strptime(conta['vencimento'],'%d/%m/%Y')
```

CRIACA√á√ÉO DE LISTAS

```python
hoje = datetime.today()

pagar = []
receber = []
pago =[] #criar lista para o pago
recebido=[] # criar lista para o recebido
contas_pagar = []
contas_receber=[]
contas_pagar_vencidas=[]
contas_receber_vencidas=[]

for conta in contas:
    if conta['tipo'] == 'Pagar':
        pagar.append(conta)

for conta in contas:
    if conta['tipo'] == 'Receber':
        receber.append(conta)

for conta in contas:
    if conta['tipo'] == 'Pagar'and conta['status'] == 'Em aberto':
        contas_pagar.append(conta)

for conta in contas:
    if conta['tipo'] == 'Receber'and conta['status'] == 'Em aberto':
        contas_receber.append(conta)

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == 'Em aberto' and conta['tipo'] == 'Pagar':
        contas_pagar_vencidas.append(conta)

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == 'Em aberto' and conta['tipo'] == 'Receber':
        contas_receber_vencidas.append(conta)            


print (f'‚ùå qtde boletos a pagar em aberto| {float(len(pagar))}')
print (f'üí≤ qtde boletos a pagar em aberto| {float(len(receber))}')

print (f'üü¢ qtde boletos a pagar em aberto| {float(len(contas_pagar))}')
print (f'üü¢ qtde boletos a receber em aberto| {float(len(contas_receber))}')    
print (f'üö® qtde boletos a pagar em aberto com atraso| {float(len(contas_pagar_vencidas))}')    
print (f'üö® qtde boletos a receber em aberto com atraso| {float(len(contas_receber_vencidas))}')    
```

CALCULOS TOTALIZADORES

```python
t_pagar =0
t_receber =0
total_pagar = 0
total_receber = 0
total_pagar_vencido =0
total_receber_vencido = 0

saldo = 0

for conta in pagar:
    t_pagar += conta['valor']

for conta in receber:
    t_receber += conta['valor']    


for conta in contas_pagar:
    total_pagar += conta['valor']

for conta in contas_receber:
    total_receber += conta['valor']

for conta in contas_pagar_vencidas:
    total_pagar_vencido += conta['valor']

for conta in contas_receber_vencidas:
    total_receber_vencido += conta['valor']
    

print(f'‚ùå contas a pagar  total| R${float(round(t_pagar,2))} ' )
print(f'üí≤ contas a receber  total| R${float(round(t_receber,2))} ' )

print(f'üü¢ contas a pagar  em aberto| R${float(round(total_pagar,2))} ' )
print(f'üü¢ contas a receber em aberto| R${float(round(total_receber,2))}' ) 
print(f'üö® contas a pagar vencidas em aberto| R${float(round(total_pagar_vencido,2))}' ) 
print(f'üö® contas a receber vencidas em aberto| R${float(round(total_receber_vencido,2))}')  

print(saldo)
```
